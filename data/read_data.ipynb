```python
import json
import os

def find_data_by_ids(jsonl_file_path, worker_id, sample_id):
    """
    æ ¹æ®worker_idå’Œsample_idæŸ¥æ‰¾å¹¶è¿”å›å¯¹åº”çš„æ•°æ®è¡Œ
    
    Args:
        jsonl_file_path (str): JSONLæ–‡ä»¶è·¯å¾„
        worker_id (int): å·¥ä½œå‘˜ID
        sample_id (int): æ ·æœ¬ID
    
    Returns:
        dict or None: æ‰¾åˆ°çš„æ•°æ®è¡Œï¼Œå¦‚æœæœªæ‰¾åˆ°åˆ™è¿”å›None
    """
    if not os.path.exists(jsonl_file_path):
        print(f"æ–‡ä»¶ä¸å­˜åœ¨: {jsonl_file_path}")
        return None
    
    with open(jsonl_file_path, 'r', encoding='utf-8') as file:
        for line_num, line in enumerate(file, 1):
            try:
                data = json.loads(line.strip())
                if data.get('worker_id') == worker_id and data.get('sample_id') == sample_id:
                    print(f"åœ¨ç¬¬ {line_num} è¡Œæ‰¾åˆ°åŒ¹é…çš„æ•°æ®")
                    return data
            except json.JSONDecodeError as e:
                print(f"ç¬¬ {line_num} è¡ŒJSONè§£æé”™è¯¯: {e}")
                continue
    
    print(f"æœªæ‰¾åˆ° worker_id={worker_id}, sample_id={sample_id} çš„æ•°æ®")
    return None

def display_data(data):
    """
    æ ¼å¼åŒ–æ˜¾ç¤ºæ•°æ®å†…å®¹
    
    Args:
        data (dict): è¦æ˜¾ç¤ºçš„æ•°æ®å­—å…¸
    """
    if data is None:
        print("æ²¡æœ‰æ•°æ®å¯æ˜¾ç¤º")
        return
    
    print("=" * 80)
    print(f"Worker ID: {data.get('worker_id')}")
    print(f"Sample ID: {data.get('sample_id')}")
    print("=" * 80)
    
    print(f"\nğŸ“ Instruction:")
    print(f"{data.get('instruction', 'N/A')}")
    
    if data.get('input'):
        print(f"\nğŸ“¥ Input:")
        print(f"{data.get('input')}")
    
    print(f"\nğŸ“„ Content:")
    print(f"{data.get('content', 'N/A')}")
    
    print(f"\nğŸ·ï¸  Label:")
    print(f"{data.get('label', 'N/A')}")
    
    print(f"\nğŸ“Š Counts: {data.get('counts', 'N/A')}")
    print("=" * 80)

# æµ‹è¯•å‡½æ•°
if __name__ == "__main__":
    # æ•°æ®æ–‡ä»¶è·¯å¾„
    jsonl_file = "merged_with_labels_and_counts.jsonl"
    
    # æµ‹è¯•æŸ¥æ‰¾åŠŸèƒ½
    print("æµ‹è¯•æŸ¥æ‰¾åŠŸèƒ½:")
    print("1. æŸ¥æ‰¾ worker_id=0, sample_id=0 çš„æ•°æ®:")
    data1 = find_data_by_ids(jsonl_file, 0, 0)
    if data1:
        display_data(data1)
    
    print("\n" + "="*100 + "\n")
    
    print("2. æŸ¥æ‰¾ worker_id=0, sample_id=1 çš„æ•°æ®:")
    data2 = find_data_by_ids(jsonl_file, 0, 1)
    if data2:
        display_data(data2)
```
